## Основы Matplotlib

### Рисунки и графики. Основная идея matplotlib

Все **Графики(Axes)**, которые мы строим в matplotlib, "живут" вутри объекта **Рисунок (Figure)**.

То есть
* **Figure (Рисунок)** является общим окном, в котором можно построить 1 или несколько **графиков (axes)**.
* **Axes** - это сам график, у которого есть оси, заголовок, легенда и т.д.

![Figure Axes Image](https://storage.googleapis.com/skytowner_public/images/uG4nzrYfQ9KDk2M87YDI/matplotlib_figure_axes_axis%20(4).png)

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid') # стиль графиков
```

Создаем Рисунок *fig* и *1 график* на нем ax, plt.subplots(1,1,1)

```python
fig, axes = plt.subplots()
```

Расчерчим рисунок сеткой 2х2 и в каждую ячеку вставим график

```python
fig, axes = plt.subplots(2, 2)
```

```python
axes # матрица из 4-х графиков
```

Добавляем по отдельности

```python
figure = plt.figure()

# figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None)
# num - номер рисунка, figsize - (float, float) ширина и высота в дюймах, dpi - разрешение, factcolor - цвет фона,
# edgecolor - цвет краев

ax1 = figure.add_subplot(2, 2, 1) # в расчерчиваем рисунок на 4 части и вставляем график в 1 ячейку
ax2 = figure.add_subplot(2, 2, 2) # вставляем график во 2 ячейку
ax3 = figure.add_subplot(2, 2, 3) # вставляем график в 3 ячейку
# 4 соответственно пустая
```

```python
fig, ax = plt.subplots()

x = [-3, -2, -1, 0, 1, 2, 3]
y = [9, 4, 1, 0, 1, 4, 9]

ax.plot(x, y) # == plt.plot(x, y) если график 1
```

### Цвет линий и стили

```python
fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0, 1, 2, 3])
y = np.array([9, 4, 1, 0, 1, 4, 9])

ax.plot(x, y)

ax.plot(x, np.sin(x), color=(1.0, 0.2, 0.3), linestyle='-')

ax.plot(x, x+5, color='blue', linestyle='--')

ax.plot(x, x+3, color='g', linestyle=':')

ax.plot(x, np.cos(x), color='0.75', linestyle='-.')

ax.plot(x, x, color='#FFDD44', linestyle='--')
```

```python
fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0, 1, 2, 3])
y = np.array([9, 4, 1, 0, 1, 4, 9])

ax.plot(x, y, marker='o')

ax.plot(0.5*x, y, marker='^')

ax.plot(2*x, y, marker='*')
```

```python
fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0, 1, 2, 3])
y = np.array([9, 4, 1, 0, 1, 4, 9])

ax.plot(x, y, color='red', linestyle='--', marker='o', alpha=0.2) # c='red', ls='--'

ax.plot(0.5*x, y, '*--g', linewidth=2.5) # lw=2.5
```

### Лимиты осей

```python
fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0, 1, 2, 3])
y = np.array([9, 4, 1, 0, 1, 4, 9])

ax.plot(x, y, c='red', ls='-', marker='o')

ax.set_xlim(0, 11)
ax.set_ylim(-5, 10)
```

### Подписи и метки

```python
fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0, 1, 2, 3])
y = np.array([9, 4, 1, 0, 1, 4, 9])

ax.plot(x, y, c='red', ls='-', marker='o', label='y=x^2')

ax.plot(0.5 * x, y, '*-g', label='y=0.5 * x^2')

ax.set_title('Заголовок нашего графика', fontsize=20)

ax.set_xlabel("Ось x")
ax.set_ylabel("Ось y")

ax.legend(loc='lower left') # loc='best'
```

```python
# plt.cla() очищаем график
# plt.clf() очищаем рисунок

fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0, 1, 2, 3])
y = np.array([9, 4, 1, 0, 1, 4, 9])

ax.plot(x, y, c='red', ls='-', marker='o', label='y=x^2')

ax.set_xticks(np.arange(-3, 3.5, 0.5))
ax.set_yticks(np.arange(0, 10, 3))

ax.spines['right'].set_visible(False)

ax.grid(color='blue', lw=2, ls='--')
# чтобы убрать сетку ax.grid()
```

## Виды графиков

### Scatter Plot

```python
fig, ax = plt.subplots()

N = 50

x = np.random.rand(N)
y = np.random.rand(N)

color = np.random.rand(N)
area = (30 * np.random.rand(N))**2

ax.scatter(x, y, s=area, c=color, alpha=0.7, marker='*', cmap='viridis')
```

### 3D Scatter Plot

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 6))

ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, s=area, c=color, alpha=0.7, marker='*', cmap='viridis')
```

### Bars

```python
fig, axes = plt.subplots(1,2)

axes[0].bar([1,2,3],[3,4,5], color='red', label='vertical bar', alpha=0.8)
axes[1].barh([0.5,1,2.5], [0,1,2])
axes[0].legend()
```

### Stacked Bar Chart

```python
fig = plt.figure(figsize=(14,10))

rect1 = plt.bar(np.arange(5), np.arange(5)**2, width=0.5, color='lightblue')
rect2 = plt.bar(np.arange(5), np.arange(5)*2, width=0.5, color='#1f77b4')

plt.legend((rect1[0], rect2[0]), ('Support', 'Freedom'))
```

### Histogram

```python
fig, ax = plt.subplots()

np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

ax.hist(x, 50, density=True, facecolor='g', alpha=0.75)
```

### Boxplot

```python
user_1 = [10, 3, 15, 21, 17, 17, 14]
user_2 = [5, 13, 10, 7, 9, 12]
data = [user_1, user_2]
fig = plt.figure(figsize=(8, 6))

ax = fig.add_axes([0, 0, 1, 1])

bp = ax.boxplot(data)

plt.xticks([1,2], ['user_1', 'user_2'])
```

### Area Chart

```python
fig, ax = plt.subplots()

x = np.arange(0, 100, 1)
y = x * 2

ax.fill_between(x, y, color='skyblue', alpha=0.2)
ax.plot(x, y, color='Slateblue', alpha=0.6)
```

### Pie Chart

```python
fig, ax = plt.subplots(figsize=(8,8))

labels = 'Jan', 'Feb', 'March', 'April', 'May', 'June'
user_1 = [10, 3, 15, 21, 17, 14]

ax.pie(user_1, labels=labels, explode=(0.07, 0, 0, 0, 0, 0), autopct='%1.1f%%', startangle=130, shadow=True)
```

### Time Series Plot

```python
import pandas as pd

plt.figure(figsize=(8,6))
ts = pd.Series(np.random.randn(100), index=pd.date_range('1/1/2020', periods=100))

ts = ts.cumsum()
ts.plot()
```

### 3D Surface

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
```
