```python
import pandas as pd
import numpy as np
```

Загрузка данных из orders.csv и customers.csv

```python
orders = pd.read_csv('orders.csv', sep=',', index_col='id')
customers = pd.read_csv('customers.csv', sep=',', index_col='id')
```

## 1. Сколько заказов, отправлено первым классом за последние 3 лет?

```python
orders['order_date'] = pd.to_datetime(orders['order_date'])
max_date = orders.query('ship_mode == "First"')['order_date'].max()
from_date = pd.Timestamp(year=max_date.year-3, month=max_date.month, day=max_date.day)
orders.query('ship_mode == "First" & order_date >= @from_date')['sales'].count()
```

**Ответ**: 645

## 2. Сколько в базе клиентов из Калифорнии?

```python
customers.query('state == "California"')['name'].count()
```

**Ответ**: 161

## 3. Сколько заказов они сделали?

```python
orders_with_customers = orders.merge(customers, how='inner', left_on='customer_id', right_index=True)
orders_with_customers.query('state == "California"')['sales'].count()
```

**Ответ**: 1006

## 4. Постройте сводную таблицу средних чеков по всем штатам за каждый год.

```python
orders_with_customers['order_date_year'] = orders_with_customers['order_date'].apply(lambda x: x.year)
orders_with_customers.pivot_table(values='sales', index='state', columns='order_date_year', aggfunc=np.mean)
```
